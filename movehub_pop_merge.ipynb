```python
# import dependencies for this notebook
import pandas as pd
```

```python
# read csv files as panda dataframes
col_df = pd.read_csv('Resources/mh_costofliving.csv')
qol_df = pd.read_csv('Resources/mh_qualityoflife.csv')
city_df = pd.read_csv('Resources/mh_cities.csv')
pop_df = pd.read_csv('Resources/Population.csv')
```

```python
# Merge the dfs together from movehub/kaggle that have the city info
big_df = col_df.join(qol_df.set_index('City'),on='City')
big_df.head()
```

```python
# merge big df with city df to get overall df for the movehub csv files found on kaggle
movehub_df = big_df.join(city_df.set_index('City'),on='City')
movehub_df
```

```python
# clean up movehub df by droping cities with unknown country(These appear as duplicates)
clean_df = movehub_df.drop_duplicates('City',keep=False)
clean_df = clean_df.dropna()
clean_df
```

```python
# Get pop df ready for merge by renaming Name column and droping unwanted columns
pop_df = pop_df.rename(columns={'Name':'City'})
pop_df.head()
pop_df = pop_df.drop('rank',1)
pop_df.head()
```

```python
# create a merged df
merged_df = clean_df.merge(pop_df, on='City', how='left')
# remove unwanted columns and drop empty rows
merged_df = merged_df.dropna().drop('Country_y',1).rename(columns={'Country_x':'Country'})
# move country to front of rows
columns = list(merged_df)
columns.insert(1,columns.pop(columns.index('Country')))
merged_df = merged_df[columns]
```

```python
# look at df
merged_df
```

```python
# save the df
merged_df.to_csv('Output/movehub_and_populations.csv',index=False)
```
